```python
from transformers import pipeline, ElectraForTokenClassification, ElectraTokenizer

test_list = ["Osman abi çukurambarda değil göksuda caddeye paralel ve ormana  bakan üç artı bir daireyi yüz elli bin TL altına almak istiyorum",
             "Hatice hanım ile caddeye bakan daireyi Nergis sokakta gördüm yüz elli bin TL ye teklif verdim",
             "arzucum ben o evi nakış sokakta gördüm hani hürriyet caddesinin sağında kalıyor",
             "abi bu evin son fiyatı nedir biz ciddi  alıcıyız",
             "geçen bi arkadaşa ziyarete gittik o sitede",
             "evet bu evimiz üç oda bir salon iki lavabosu var",
             "3+1 olmaz da 2+1 belki olur"]
```

```python
model_try = ElectraForTokenClassification.from_pretrained("/content/trained_model")
tokenizer_try = ElectraTokenizer.from_pretrained("/content/trained_model")
ner_try=pipeline('ner', model=model_try, tokenizer=tokenizer_try, grouped_entities=False)
print("try:")
for text in test_list:
    print(text, ": ", ner_try(text), "\n")
```

```python
model_try = ElectraForTokenClassification.from_pretrained("/content/trained_model2")
tokenizer_try = ElectraTokenizer.from_pretrained("/content/trained_model2")
ner_try=pipeline('ner', model=model_try, tokenizer=tokenizer_try, grouped_entities=False)
print("try:")
for text in test_list:
    print(text, ": ", ner_try(text), "\n")
```
